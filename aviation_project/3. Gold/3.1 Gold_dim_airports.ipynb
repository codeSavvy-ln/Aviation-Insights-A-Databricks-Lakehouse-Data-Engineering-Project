```python
%sql
Create or Replace table aviation_project.gold.dim_airports as
SELECT
    sha2(concat_ws('||', IATA_CODE), 256) AS AIRPORTS_SK,
    IATA_CODE,
    CITY,
    STATE,
    COUNTRY,
    LATITUDE,
    LONGITUDE,
    AIRPORT_REGION as REGION,
    AIRPORT_CONTINENT AS CONTINENT_GROUP,
    AIRPORT_SIZE AS AIRPORT_SIZE_CATEGORY,
    LOCATION_VALIDATION

FROM aviation_project.silver.airports
```

**Data Quality Checks:**
 - Primary Key uniqueness
 - Null values
 - Iata_code validity

```python
%sql
-- primary key uniqueness
select airports_sk, count(airports_sk) as dup
from aviation_project.gold.dim_airports
group by airports_sk
having count(airports_sk) > 1
```

```python
%sql
SELECT
  COUNT(*) AS total_records,

  SUM(CASE WHEN AIRPORTS_SK IS NULL THEN 1 ELSE 0 END) AS AIRPORTS_SK_NULLS,
  SUM(CASE WHEN IATA_CODE IS NULL THEN 1 ELSE 0 END) AS IATA_CODE_NULLS,
  SUM(CASE WHEN CITY IS NULL THEN 1 ELSE 0 END) AS CITY_NULLS,
  SUM(CASE WHEN STATE IS NULL THEN 1 ELSE 0 END) AS STATE_NULLS,
  SUM(CASE WHEN COUNTRY IS NULL THEN 1 ELSE 0 END) AS COUNTRY_NULLS,
  SUM(CASE WHEN LATITUDE IS NULL THEN 1 ELSE 0 END) AS LATITUDE_NULLS,
  SUM(CASE WHEN LONGITUDE IS NULL THEN 1 ELSE 0 END) AS LONGITUDE_NULLS,
  SUM(CASE WHEN REGION IS NULL THEN 1 ELSE 0 END) AS REGION_NULLS,
  SUM(CASE WHEN CONTINENT_GROUP IS NULL THEN 1 ELSE 0 END) AS CONTINENT_GROUP_NULLS,
  SUM(CASE WHEN AIRPORT_SIZE_CATEGORY IS NULL THEN 1 ELSE 0 END) AS AIRPORT_SIZE_CATEGORY_NULLS,
  SUM(CASE WHEN LOCATION_VALIDATION IS NULL THEN 1 ELSE 0 END) AS LOCATION_VALIDATION_NULLS
FROM aviation_project.gold.dim_airports;
```
