‚úÖ 1. Top 10 Busiest Routes (Altair ‚Üí Matplotlib)

```python
 busiest = (
        filtered.groupby(['ORIGIN_AIRPORT_SK','DESTINATION_AIRPORT_SK'])
        .size()
        .reset_index(name='FLIGHT_COUNT')
        .sort_values(by='FLIGHT_COUNT', ascending=False)
        .head(10)
    )
```

```python
st.subheader("‚úàÔ∏è Top 10 Busiest Routes")

plt.figure(figsize=(12, 6))
plt.bar(
    busiest["ORIGIN_CODE"],
    busiest["FLIGHT_COUNT"]
)

plt.xlabel("Origin Airport Code")
plt.ylabel("Flight Count")
plt.title("Top 10 Busiest Routes")

plt.xticks(rotation=45)
st.pyplot(plt)
plt.clf()
```
